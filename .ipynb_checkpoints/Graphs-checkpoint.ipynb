```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
print(os.listdir("."))
data = np.loadtxt("gridsize_results", delimiter=",")
```

```python
print(data)
```

```python
plt.figure()
plt.plot(data[:,0], data[:,1], marker = "x")
plt.title("Time versus Environment Size")
plt.ylabel("Time for model checking / s")
plt.xlabel("Grid axis size")
print(data[:,0])
print(data[:,1])
```
